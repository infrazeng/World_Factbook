```python
%%sql 
SELECT *
FROM facts
LIMIT 5;
```

```python
%%sql
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts;
```

```python
%%sql 
SELECT name
FROM facts
ORDER BY population
LIMIT 1;
```

```python
%%sql 
SELECT *
FROM facts
WHERE population=0;
```

```python
%%sql 
SELECT *
FROM facts
ORDER BY population DESC
LIMIT 1;
```

```python
%%sql 
SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
WHERE (name!='World');
```

```python
%%sql 
SELECT AVG(population), AVG(area)
FROM facts
WHERE (name!='World');
```

```python
%%sql
SELECT *
FROM facts
WHERE (population > (SELECT AVG(population) FROM facts))
AND (area < (SELECT AVG(area) FROM facts));
```

```python
%%sql
SELECT name
FROM facts
WHERE name!='World'
ORDER BY population DESC
LIMIT 1;
```

```python
%%sql
SELECT name
FROM facts
WHERE name!='World'
ORDER BY population_growth DESC
LIMIT 1;
```

```python
%%sql
SELECT *
FROM facts
WHERE (name!='World')
AND (area_water > area_land);
```

```python
%%sql 
SELECT name, CAST(area_water AS Float) / area_land AS water_land_ratio 
FROM facts
WHERE name!='World'
ORDER BY water_land_ratio DESC
LIMIT 5;
```

```python
%%sql 
SELECT name, population_growth
FROM facts
ORDER BY population_growth DESC
LIMIT 5;
```

```python
%%sql 
SELECT name
FROM facts
WHERE death_rate > birth_rate;
```

```python
%%sql
SELECT name, (CAST(population as Float) / area
FROM facts
ORDER BY population / area
LIMIT 5;
```
